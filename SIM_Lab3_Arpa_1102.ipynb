```python
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Generate 1×10 random numbers in [0,1)
random_numbers = np.random.rand(10)

# Plot the histogram
plt.hist(random_numbers, bins=5, alpha=0.5, color='blue', edgecolor='black')

# Labels and title
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Histogram of Uniformly Distributed Random Numbers')
plt.grid(True)

# Show the plot
plt.show()
```

```python
# Generate 500 random values from a normal distribution with mean=0 and std=1
NormalValues = np.random.normal(0, 1, 500)

# Create a histogram
plt.hist(NormalValues, bins=15, edgecolor='black')
plt.title('Histogram of Normal Distribution (mean=0, std=1)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python

non_seeded_matrix = np.random.randint(1, 50, (3, 3))
print("\nNon-Seeded Random Matrix:\n", non_seeded_matrix)

# With Seed
np.random.seed(10)  # Setting seed to 10
seeded_matrix = np.random.randint(1, 50, (3, 3))
print("Seeded Random Matrix:\n", seeded_matrix)

# Without Seed
np.random.seed(None)  # No seed, generates different values each time
non_seeded_matrix = np.random.randint(1, 50, (3, 3))
print("\nNon-Seeded Random Matrix:\n", non_seeded_matrix)

# Explanation:
# The matrix generated with the fixed seed (np.random.seed(10)) will remain the same
# every time the script is run. The matrices generated without a seed will change on
# each run, demonstrating the non-reproducible nature of unseeded random generation.
```

```python
import random  # to create random numbers

def estimate_pi(num_points):
    inside = 0  # how many points inside the circle

    for _ in range(num_points):
        x = random.random()  # pick a random number from 0 to 1
        y = random.random()
        if x*x + y*y <= 10*10:
            inside += 1  # it's inside the circle

    pi = (400 * inside) / (100 * num_points)
    print("Estimated pi for", num_points, "points is:", pi)

# Try this with different number of points
estimate_pi(2200)

# MY data
student_id_last_two = "02"
full_name = "Arpa Bhowmik"
```
