```python
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns

scalar = 3.7
arr = np.array([1, 2, 3, 4, 5, 6])

sin_values = np.sin(arr)
cos_values = np.cos(arr)
tan_values = np.tan(arr)
asin_values = np.arcsin(np.clip(arr / np.max(arr), -1, 1))
acos_values = np.arccos(np.clip(arr / np.max(arr), -1, 1))
atan_values = np.arctan(arr)

print("sin:", sin_values)
print("cos:", cos_values)
print("tan:", tan_values)
print("asin:", asin_values)
print("acos:", acos_values)
print("atan:", atan_values)
```

```python

vec_matrix = np.array([[2.1, 4.7, -1.2],
                       [3.3, 0.5, -2.8]])

print("\nOriginal 2x3 matrix:\n", vec_matrix)

# Max and Min with indices
print("Max value:", np.max(vec_matrix))
print("Index of max:", np.unravel_index(np.argmax(vec_matrix), vec_matrix.shape))
print("Min value:", np.min(vec_matrix))
print("Index of min:", np.unravel_index(np.argmin(vec_matrix), vec_matrix.shape))

# Vector length
print("Vector length (norm):", np.linalg.norm(vec_matrix))

# Sorted versions
print("Row-wise sorted (descending):\n", -np.sort(-vec_matrix, axis=1))
print("Entire matrix sorted (descending):\n", np.sort(vec_matrix.flatten())[::-1])

# Sum and Product
print("Sum of elements:", np.sum(vec_matrix))
print("Product of elements:", np.prod(vec_matrix))

# Median, Mean, Standard Deviation
print("Median:", np.median(vec_matrix))
print("Mean:", np.mean(vec_matrix))
print("Standard Deviation:", np.std(vec_matrix))


# Bonus: Heatmap & Random Input

# Random 3x3 matrix using uniform distribution
rand_matrix = np.random.uniform(-5, 5, (3, 3))
print("\nRandom 3x3 matrix:\n", rand_matrix)

# Heatmap visualization
plt.figure(figsize=(6, 4))
sns.heatmap(rand_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Heatmap of Random Matrix")
plt.show()
```
