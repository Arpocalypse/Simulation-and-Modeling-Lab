```python
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Create two 3Ã—3 matrices with random integers
A = np.random.randint(1, 11, size=(3, 3))
B = np.random.randint(1, 11, size=(3, 3))

print("Matrix A:\n", A)
print("\nMatrix B:\n", B)

# Matrix addition
C = A + B
print("\nThe sum of A and B is:\n", C)

# Matrix subtraction
D = A - B
print("\nThe difference of A and B is:\n", D)

# Matrix multiplication
E = np.matmul(A, B)  # or A @ B
print("\nThe product of A and B is:\n", E)

# Determinant of matrix A
det_A = np.linalg.det(A)
print(f"\nDeterminant of A: {det_A:.2f}")

# Inverse of matrix A
if det_A != 0:
    inv_A = np.linalg.inv(A)
    print("\nInverse of A:\n", inv_A)
else:
    print("\nMatrix A is singular, inverse cannot be computed.")

# Rank of matrix A
rank_A = np.linalg.matrix_rank(A)
print(f"\nRank of A: {rank_A}")
```

```python

# Create and plot two random vectors (length 15)
A = np.random.rand(15)
B = np.random.rand(15)

# Create an index for x-axis
x = np.arange(len(A))  # length of vector A

print(x)

# Plot the data with x and y values
plt.plot(x, A, marker='o', linestyle='-', label="Vector A")
plt.plot(x, B, marker='s', linestyle='--', label="Vector B")

# Labels and title
plt.xlabel("Index")
plt.ylabel("Value")
plt.title("Line Graph of Two Random Vectors")
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
# Generate a 4x4 matrix with random values
matrix = np.random.rand(4, 4)
print(matrix)
# Plot the heatmap
sns.heatmap(matrix, annot=True, cmap='coolwarm', linewidths=0.5, linecolor ='black')
plt.title('Heatmap of 4x4 Matrix')
plt.xlabel("Columns")
plt.ylabel("Rows")
plt.show()
```

```python
# Create two random 4x4 matrices with integer values between 0 and 17
A = np.random.randint(0, 10, size=(4, 4))  # Random integers between 0 and 9
B = np.random.randint(0, 10, size=(4, 4))  # Random integers between 0 and 9

print("Matrix A:\n", A)
print("Matrix B:\n", B)

# Matrix operations
matrix_sum = A + B  # Matrix addition
matrix_diff = A - B  # Matrix subtraction
matrix_prod = np.dot(A, B)  # Matrix multiplication (dot product)

# Bar plot to visualize results
fig, axes = plt.subplots(1, 3, figsize=(15, 5)) #figure object, subplot array

# Sum
axes[0].bar(range(1, 17), matrix_sum.flatten(), color='b')
axes[0].set_title('Matrix Addition')
axes[0].set_xlabel('Index')
axes[0].set_ylabel('Value')

# Difference
axes[1].bar(range(1, 17), matrix_diff.flatten(), color='g')
axes[1].set_title('Matrix Subtraction')
axes[1].set_xlabel('Index')
axes[1].set_ylabel('Value')

# Product
axes[2].bar(range(1, 17), matrix_prod.flatten(), color='r')
axes[2].set_title('Matrix Multiplication')
axes[2].set_xlabel('Index')
axes[2].set_ylabel('Value')

plt.tight_layout() #plt.tight_layout() is a Matplotlib function that automatically adjusts the layout of subplots and figures to prevent overlapping and ensure a visually appealing presentation. It modifies the spacing between subplots, axes, labels, and titles to improve readability.
plt.show()
```
